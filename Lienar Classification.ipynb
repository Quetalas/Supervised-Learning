```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from matplotlib.colors import ListedColormap
```

```python
cmap = ListedColormap(['blue','green'])
X, y = make_classification(class_sep=0.80,n_samples=100, n_features=2, n_classes=2, n_redundant=0, n_repeated=0)
plt.scatter(X[:,0], X[:,1], cmap=cmap, c=y)
```

```python
from sklearn.cross_validation import train_test_split
cmap2 = ListedColormap(['red','yellow'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
fig, axes = plt.subplots(1, 2, figsize=(10,3))
for model, ax in zip([LinearSVC(), LogisticRegression()], axes):
    clf = model.fit(X_train, y_train)
    ax.scatter(X_train[:,0], X_train[:,1], cmap=cmap, c=y_train)
    ax.scatter(X_test[:,0], X_test[:,1], cmap=cmap2, c=y_test)
    f = lambda x: clf.coef_[0,0]*x/clf.coef_[0,1] - clf.intercept_/clf.coef_[0,1]
    x = np.arange(-3,3)
    ax.plot(x, f(x))
    ax.set_ylim(-3,3)
    print(clf.score(X_test, y_test))
```
