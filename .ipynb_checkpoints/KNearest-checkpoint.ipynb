```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
from sklearn.cross_validation import train_test_split
from matplotlib.colors import ListedColormap
%matplotlib inline
```

```python
from ipywidgets import interact, IntSlider
```

```python
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

iris = load_iris()
X = iris.data
y = iris.target

x_min, x_max = X[:, 1].min() - 1, X[:, 1].max() + 1
y_min, y_max = X[:, 2].min() - 1, X[:, 2].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                         np.arange(y_min, y_max, 0.02))
z = np.c_[xx.ravel(), yy.ravel()]

X_train, X_test, y_train, y_test = train_test_split(X[:,[1,2]], y)
train_accuracy = []
test_accuracy = []
for k in range(1, 11):    
    
    knn = KNeighborsClassifier(n_neighbors=k).fit(X_train, y_train)
    test_accuracy.append(knn.score(X_test, y_test))
    train_accuracy.append(knn.score(X_train, y_train))
def f(k):
    train_accuracy = []
    knn = KNeighborsClassifier(n_neighbors=k).fit(X_train, y_train)
    pred = knn.predict(z)
    pred = pred.reshape(xx.shape)
    plt.pcolormesh(xx, yy, pred, cmap=cmap_light)
    plt.scatter(X[:,1], X[:,2], cmap=cmap_bold, c =iris.target)
    plt.show()    
    
```

```python
interact(f, k=IntSlider(min=1, max = 112, step=1, value=0))
```

```python
plt.plot(range(1,11), test_accuracy,c='g')
plt.plot(range(1,11), train_accuracy)
plt.xlabel('Количество соседей')
plt.ylabel('Точность предсказания')
plt.legend(['На тестовой выборке','на обучающей выборке'])
plt.show()
```

```python
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

iris = load_iris()
X = iris.data
y = iris.target

x_min, x_max = X[:, 1].min() - 1, X[:, 1].max() + 1
y_min, y_max = X[:, 2].min() - 1, X[:, 2].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                         np.arange(y_min, y_max, 0.02))
z = np.c_[xx.ravel(), yy.ravel()]

X_train, X_test, y_train, y_test = train_test_split(X[:,[1,2]], y)
train_accuracy = []
test_accuracy = []
for k in range(1, 112):    
    
    knn = KNeighborsClassifier(n_neighbors=k,weights='distance').fit(X_train, y_train)
    test_accuracy.append(knn.score(X_test, y_test))
    train_accuracy.append(knn.score(X_train, y_train))
def f(k):
    train_accuracy = []
    knn = KNeighborsClassifier(n_neighbors=k,weights='distance').fit(X_train, y_train)
    pred = knn.predict(z)
    pred = pred.reshape(xx.shape)
    plt.pcolormesh(xx, yy, pred, cmap=cmap_light)
    plt.scatter(X[:,1], X[:,2], cmap=cmap_bold, c =iris.target)
    plt.show()    
interact(f, k=IntSlider(min=1, max = 112, step=1, value=0))
```

```python
plt.plot(range(1,112), test_accuracy,c='g')
plt.plot(range(1,112), train_accuracy)
plt.xlabel('Количество соседей')
plt.ylabel('Точность предсказания')
plt.legend(['На тестовой выборке','на обучающей выборке'])
plt.show()
```
